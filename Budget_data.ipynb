```python
import pandas as pd
from pathlib import Path
```

```python
Path.cwd()
```

```python
csvpath = Path("../Resources/budget_data.csv")
```

```python
budget_data = pd.read_csv(csvpath)
budget_data.head()
```

```python
# The total number of months included in the dataset
Months = budget_data['Date'].count()
total_months = f'Total Months:{Months}'
total_months
```

```python
# The net total amount of Profit/Losses over the entire period.
net_profit = budget_data['Profit/Losses'].sum()
net_profit
```

```python
# The average of the changes in Profit/Losses over the entire period
budget_data['Profit/Losses'].sum()
budget_data['Change'] = budget_data['Profit/Losses'] - budget_data['Profit/Losses'].shift(1)
budget_data
```

```python
# The greatest increase in profits (date and amount) over the entire period
column = (budget_data['Profit/Losses'] - budget_data['Profit/Losses'].shift(1)).mean()
column
```

```python
column_max = (budget_data['Profit/Losses'] - budget_data['Profit/Losses'].shift(1)).max()
column_max
```

```python
column_min = (budget_data['Profit/Losses'] - budget_data['Profit/Losses'].shift(1)).min()
column_min
```

```python
print(f'Total Months: {total_months}')
print(f'Net Profit: {net_profit}')
print(f'Average of changes: {column}')
print(f'Greatest Increase: {column_max}')
print(f'Greatest Decrease: {column_min}')
```
